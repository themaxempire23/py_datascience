imports


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

data loading


```python
tips = sns.load_dataset('tips')
```

dataset summary


```python
# describing the dataset
print(tips.describe())
```

```python
# finding the mean, median and mode
print("mean tip:", tips['tip'].mean())
print("mean tip:", tips['tip'].median())
print("mean tip:", tips['tip'].mode()[0])
```

finding the varience, standard deviation


```python
# finding the stadard deviation
# std ->standard deviation, var -> variable

tip_std = tips['tip'].std()
tip_var = tips['tip'].var()
tip_range = tips['tip'].max() - tips['tip'].min()

print(tip_std)
print(tip_var)
print(tip_range)
```

visualing the statistics and distribution


```python
sns.histplot(tips['tip'], kde=True, color='skyblue')
plt.title('distribution of tip amounts')
plt.xlabel('Tip amaount (N$)')
plt.ylabel('frequency')
plt.show()
```
