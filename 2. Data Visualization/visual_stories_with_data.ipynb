imports


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

data loading


```python
sns.set_theme()
tips = sns.load_dataset('tips')
print(tips.head(5))
```

Exploring the data


```python
# using a histplot
sns.histplot(data=tips, x='total_bill', hue='sex', kde=True)
plt.title('distribution of total bills by gender')
plt.xlabel('count')
plt.show()
```

finding the figure size


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker')
plt.title('tips vs total bill by gender and smoking status')
plt.xlabel('total bill')
plt.ylabel('tip amount')
plt.legend(title='Gender / Smoker')
plt.grid(True)
plt.show()
```
